```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
api_key = os.getenv("GROQ_API_KEY")
model = ChatGroq(
    api_key=api_key,
    model = "llama3-70b-8192"
)
model
```

```python
from langchain.schema import(
    AIMessage,
    HumanMessage,
    SystemMessage
)
```

```python
speech = '''
I would like to speak about Modi's vision for Viksit Bharat — a developed India.
Under Prime Minister Narendra Modi's leadership, the nation is working towards becoming a self-reliant, technologically advanced, and economically strong country by 2047, which marks 100 years of independence. The goal of Viksit Bharat is not just development in infrastructure, but also inclusive growth — empowering youth, supporting farmers, fostering innovation, and ensuring sustainability. This vision aims to transform India into a global leader while preserving our cultural values and democratic principles.
'''
```

```python
speech
```

```python
chat_messages = [
    SystemMessage(content ="You are a expert in summarizing speeches"),
    HumanMessage(content = f"Please summarize the following speech: \n Text:{speech}"),
]
```

```python
model.get_num_tokens(speech)
```

```python
## getting the summary
model(chat_messages)
```

```python
## Prompt Template Text Summarization
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

generic_template = """
Write a concise summary of the following speech:
Speech:{speech}
Translate the summary into {language} language.

"""

prompt = PromptTemplate(
    input_variables=['speech', 'language'],
    template=generic_template
)

prompt
```

```python
completed_prompt = prompt.format(speech=speech,language='French')
```

```python
model.get_num_tokens(completed_prompt)
```

```python
llm_chain = LLMChain(
    llm=model,
    prompt=prompt,)
summary = llm_chain.run({"speech":speech,"language":"Telugu"})
summary
```

StuffDocument Chain Summarization

```python
from langchain_community.document_loaders import PyPDFLoader

loader =PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split() 
docs
```

```python
template ="""
Write a concise summary of the following speech:
Speech:{text}

"""
prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain  = load_summarize_chain(
    llm=model,
    chain_type="stuff",
    prompt=prompt,
    verbose=True
)
output_summary = chain.run(docs)
output_summary
```

### Map Reduce Summarize Large Documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_documnets = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=100,
).split_documents(docs)

final_documnets
```

```python
len(final_documnets)
```

```python
prompt
```

```python
chunks_prompts ="""
Write a concise summary of the following speech:
Speech:'{text}'
Summary:
"""
map_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=chunks_prompts
)

```

```python
final_prompt = """
Provide the final summary of the entire speech with these important points.
Add a Motivation Title and start a with a introduction and provide the summary in numerical points
for the speech.
speech:{text}

"""

final_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=final_prompt
)

final_prompt_template
```

```python
summary_chain = load_summarize_chain(
    llm=model,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
   
    verbose=True
)
output_summary = summary_chain.run(final_documnets)
output_summary
```

## Refine Chain for summarization

```python
chain = load_summarize_chain(
    llm=model,
    chain_type="refine",
    verbose=True
)
output_summary = chain.run(final_documnets)
output_summary
```
