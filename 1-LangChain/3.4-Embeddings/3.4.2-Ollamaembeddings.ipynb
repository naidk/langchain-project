### Ollama

Ollama support embeddings models ,making it possible to retrival augmented Genearation (RAG) applications that combine text prompts with existing documnets or other data

```python
from langchain_community.embeddings import OllamaEmbeddings

```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")
    ) # by default uses the "llama2" model
embeddings
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of the Greek alphabet.",
        "Beta is the second letter of the Greek alphabet.",
    ]
)
```

```python
r1
```

```python
len(r1[0]) # length of the embedding vector
```

```python
embeddings.embed_query("What is the second letter of the Greek alphabet?")
```
